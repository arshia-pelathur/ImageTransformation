```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('C:/Users/ArshiaPelathur/Desktop/StrataScratch/Transforming_Image/img1.jpg',1)         # in BGR format by default
```

```python
fig,ax = plt.subplots(3,2,figsize=(8,8))

RGB_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.subplot(3, 2, 1)
plt.imshow(RGB_image)
ax[0,0].set_title('Original')

plt.subplot(3, 2, 2)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image,cmap='gray')
ax[0,1].set_title('GrayScale')

plt.subplot(3, 2, 3)
inverted_image = cv2.bitwise_not(gray_image)
plt.imshow(inverted_image,cmap='gray')
ax[1,0].set_title('Inverted')

plt.subplot(3, 2, 4)
blurred_image = cv2.GaussianBlur(inverted_image, (21,21), 0)
plt.imshow(blurred_image,cmap='gray')
ax[1,1].set_title('Blurred')

plt.subplot(3, 2, 5)
invblurred_image = cv2.bitwise_not(blurred_image)
plt.imshow(invblurred_image,cmap='gray')
ax[2,0].set_title('Inverted Blurred')

plt.subplot(3, 2, 6)
pencil_sketch = cv2.divide(gray_image, invblurred_image, scale=256.0)
plt.imshow(pencil_sketch,cmap='gray')
ax[2,1].set_title('Pencil Sketch')

for i in range(3):
    for j in range(2):
        ax[i,j].axis('Off')

plt.tight_layout()
plt.show()
```
