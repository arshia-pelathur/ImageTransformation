```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('C:/Users/ArshiaPelathur/Desktop/StrataScratch/Transforming_Image/img1.jpg',1)         # in BGR format by default
RGB_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.subplot(1,3,1)
plt.imshow(RGB_image)
# cv2.waitKey(0)

plt.subplot(1,3,2)
gray_image = cv2.cvtColor(RGB_image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image)
# cv2.waitKey(0)

plt.subplot(1,3,3)
inverted_image = cv2.bitwise_not(gray_image)
plt.imshow(inverted_image)
# cv2.waitKey(0)

plt.subplot(2,3,1)
blurred_image = cv2.GaussianBlur(inverted_image, (21,21), 0)
plt.imshow(blurred_image)
# cv2.waitKey(0)

plt.subplot(2,3,2)
invblurred_image = cv2.bitwise_not(blurred_image)
plt.imshow(invblurred_image)
# cv2.waitKey(0)

pencil_sketch = cv2.divide(gray_image, invblurred_image, scale=256.0)
plt.imshow(pencil_sketch)
# cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(14,8))

plt.subplot(1,2,1)
RGB_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.title('Original image', size=18)
plt.imshow(RGB_img)

plt.axis('off')

plt.subplot(1,2,2)
plt.title('Sketch', size=18)
rgb_pencil_sketch=cv2.cvtColor(pencil_sketch, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_pencil_sketch)
plt.axis('off')
plt.show()
```
